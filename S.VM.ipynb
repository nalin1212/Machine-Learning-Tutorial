# Hyperplane

A hyperplane is a flat affine subspace that has one dimension less than its ambient space.

In 2D space, a hyperplane is a line.

In 3D space, a hyperplane is a plane.

In n-dimensional space, a hyperplane is an (n−1)-dimensional subspace.

# 🔹 Hyperplane in Machine Learning:
In algorithms like Support Vector Machines (SVMs):

The hyperplane is used to separate different classes in classification tasks.

The best hyperplane is the one that maximizes the margin between classes.

Example in 2D:

If we have two classes of points, a line (hyperplane in 2D) can be drawn to separate them.



```python
from sklearn.datasets import load_breast_cancer
df=load_breast_cancer()
print(df.DESCR) 
```

```python
x= df.data 
y=df.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
train_sc=scaler.fit_transform(X_train)
test_sc=scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC( kernel='rbf',C=1.0,gamma='scale')
model.fit(train_sc,y_train)
```

```python
y_pred =model.predict(test_sc)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix ,classification_report
acc=accuracy_score (y_test,y_pred)
cm= confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
```

```python
print(acc,cm,cr)
```

🔹 Why Use a Kernel?
Because some data that is not linearly separable in its original space might become linearly separable in a higher-dimensional space.

Instead of manually transforming the data, you use a kernel function to implicitly map the data into a higher-dimensional space.

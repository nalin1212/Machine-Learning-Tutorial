# Machine Learning

It is a sub feild of of A.I . where machine learn from historical data 


---

## 📦 Supervised Learning
- **Definition:** Learns from labeled data.
- **Examples:** Regression, Classification  
- **Popular Algorithms:**
  - Linear Regression
  - Logistic Regression
  - Decision Tree
  - Random Forest
  - Support Vector Machine (SVM)
  - k-Nearest Neighbors (k-NN)
  - Naive Bayes
  - Neural Networks

---

## 🎲 Unsupervised Learning
- **Definition:** Learns patterns from unlabeled data.
- **Examples:** Clustering, Dimensionality Reduction  
- **Popular Algorithms:**
  - K-Means Clustering
  - DBSCAN
  - Hierarchical Clustering
  - Principal Component Analysis (PCA)
  - Independent Component Analysis (ICA)
  - Autoencoders

---

## 🎮 Reinforcement Learning
- **Definition:** Agent interacts with an environment to maximize cumulative reward.
- **Examples:** Game AI, Robotics, Self-driving cars  
- **Popular Algorithms:**
  - Q-Learning
  - SARSA
  - Deep Q-Networks (DQN)
  - Policy Gradient
  - Actor-Critic Methods

---

## 🔁 Other Types

### Semi-Supervised Learning
- Combines small labeled data + large unlabeled data.
- Example: Google Photos face grouping.

### Self-Supervised Learning
- Learns from unlabeled data by generating pseudo-labels.
- Used in:
  - NLP: BERT
  - Vision: SimCLR, MoCo

---

## 🤖 Deep Learning
- Subfield of ML using multi-layer neural networks.
- **Popular Architectures:**
  - CNNs (Convolutional Neural Networks) — for images
  - RNNs/LSTMs — for sequences (text/time-series)
  - Transformers — for NLP, vision (e.g., GPT, BERT)

---
